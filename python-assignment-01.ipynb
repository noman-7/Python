**Write a program that checks if a given number is prime or not**

```python
n = int(input())
flag = False

if n == 1:
    print(n, "is not a prime number")
elif n > 1:
    for i in range(2, n):
        if (n % i) == 0:
            flag = True
            break

    if flag:
        print(n, "is not a prime number")
    else:
        print(n, "is a prime number")
```

**Write a program that finds the sum of all even numbers between 1 and `n`**

```python
n = int(input())
c = 2
s = 0

while c <= n:
    if c%2 == 0:
      s = s + c
    c = c + 1  
print(s)
```

**Write a program that reverses a given number**

```python
x = int(input())
y = 0

while x > 0:
    digit = x % 10
    y = y * 10 + digit
    x = x // 10

print("Reversed Number: " + str(y))
```

**Write a program that checks if a given string, is a palindrome**

```python
x = input()
y = x[::-1]

while x == y:
    print("Yes")
    break
else:
    print("No")
```

**Write a program that generates a random number and allows the user to guess it**

```python
#collected from ChatGPT

import random

# Generate a random number between 1 and 100
random_number = random.randint(1, 100)

print("I've selected a number between 1 and 100. Try to guess it!")

# Loop until the user guesses correctly
while True:
    # Get user's guess
    user_guess = int(input("Enter your guess: "))
    
    # Check the guess
    if user_guess < random_number:
        print("Too low! Try again.")
    elif user_guess > random_number:
        print("Too high! Try again.")
    else:
        print("Congratulations! You've guessed the number.")
        break
```

**Write a program that finds the greatest common divisor (GCD) of two numbers**

```python
# Get user input for two numbers
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

# Initialize variables
a = num1
b = num2

# Euclidean Algorithm to find GCD
while b != 0:
    a, b = b, a % b

# Print the GCD
print(f"The greatest common divisor (GCD) of {num1} and {num2} is: {a}")
```
